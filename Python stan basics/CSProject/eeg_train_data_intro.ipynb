```python
eeg_sample_id = 568657#3445985682#1982923927#
train_data = pd.read_csv('data/train.csv')
print(train_data.shape)
spec_sample_id = train_data.index[train_data['eeg_id'] == eeg_sample_id][0]
```

```python
file_path = 'data/train_eegs/'+str(eeg_sample_id)+'.parquet'
print(file_path)
data = pd.read_parquet(file_path, engine='pyarrow')
```

```python
cols = data.columns
print(cols)
```

```python
import matplotlib.pyplot as plt

col = 'Fp1'
plt.figure(figsize=(15, 1))
plt.plot(np.arange(1,len(data[col])+1), data[col], linewidth=0.3)
plt.xlabel('epoch')
plt.ylabel('frequency')
#plt.ylim(-500,500)
plt.show()
```

```python
col = 'F3'
plt.figure(figsize=(15, 1))
plt.plot(np.arange(1,len(data[col])+1), data[col], linewidth=0.1)
plt.xlabel('epoch')
plt.ylabel('frequency')
#plt.ylim(-500,500)
plt.show()
```
